```python
def GetLegendre(n):

  x = sym.Symbol('x',Real=True)
  y = sym.Symbol('y',Real=True)

  y = (x**2 - 1)**n

  p = sym.diff(y,x,n)/(2**n * math.factorial(n))

  return p

for i in range(20):
  print(GetLegendre(i))




for i in range(1, 21):
    zeros, weights = roots_legendre(i)
    print(f'Polinomio de Legendre P_{i}:')
    print(f'  Ceros: {zeros}')


Roots1, Weights1 = np.polynomial.legendre.leggauss(20)
print(Roots1,Weights1)
```
