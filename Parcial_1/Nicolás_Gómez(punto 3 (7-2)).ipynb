```python
import matplotlib.pyplot as plt
import numpy as np

# Valores de x y y
x_values = [0, np.pi/3, np.pi/2]
y_values = [np.sin(0), np.sin(np.pi/3), np.sin(np.pi/2)]

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, 'o', color='blue')  # 'o' para los puntos
plt.title('Conjunto de Soporte Ω')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.xticks([0, np.pi/3, np.pi/2], ['0', 'π/3', 'π/2'])
plt.yticks([0, np.sin(np.pi/3), 1], ['0', 'sin(π/3)', '1'])

# Etiquetar los puntos
for (x, y) in zip(x_values, y_values):
    plt.text(x, y, f'({x:.2f}, {y:.2f})', fontsize=12, ha='right')

plt.grid(True)
plt.show()
```

```python
# Valores de x y y
X = [0, np.pi/3, np.pi/2]
Y = [np.sin(0), np.sin(np.pi/3), np.sin(np.pi/2)]
x1= np.linspace(0,(np.pi/2),100)

def Lagrange(x, X, i):
  L=1
  for j in range(len(X)):
    if j!=i:
      L*= (x-X[j])/(X[i]-X[j])
  return L

def Interpolate(x,X,Y):
  L = 0
  for i in range(len(X)):
    L += Y[i]*Lagrange(x,X,i)
  return L

y1=Interpolate(x1,X,Y)

plt.scatter(X,Y,color='r')
plt.plot(x1,y1)





```
