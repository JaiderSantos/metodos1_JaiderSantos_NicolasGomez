```python
h2=0.001
X=np.linspace(-4,4,25)
x,y=np.meshgrid(X,X)
def Pow_De_Flujo(x,y):
  F=2*x*(1-(4/(x**2+y**2)))
  return F
def  dev_y(x,y,F,h):
  dy= (F(x,y+h)-F(x,y-h))/h*2
  return dy
def dev_x(x,y,F,h):
  dx= (F(x+h,y)-F(x-h,y))/h*2
  return dx
```

```python
figura=plt.figure(1,figsize=(10,10))
ax=figura.add_subplot(111)
for i in range(len(X)):
  for j in range(len(X)):
    if np.sqrt(X[i]**2+X[j]**2) > 2:
      ax.quiver(X[i],X[j],dev_x(X[i],X[j],Pow_De_Flujo,h2),-dev_y(X[i],X[j],Pow_De_Flujo,h2))
```

```python
x=np.linspace(-10,10,100)
def func(x):
  f=1/(np.sqrt(1+np.exp(x**2)))

  return f
def derivadas(f,x,h):
  return (f(x+h)-f(x-h))/(2*h)

plt.plot(x,func(x))
plt.plot(x,derivadas(func,x,0.05))
```
