Generales de Probabilidad

```python
#punto1
# Número total de casos en el espacio muestral (6 * 6 = 36)
total_casos = 36

# Casos favorables para A (suma <= 3)
casos_A = 3  # (1,1), (1,2), (2,1)

# Casos favorables para B (primer lanzamiento impar)
casos_B = 18  # 3 resultados para el primer lanzamiento * 6 para el segundo

# Casos favorables para A ∩ B (suma <= 3 y primer lanzamiento impar)
casos_A_interseccion_B = 2  # (1,1), (1,2)

# Probabilidad de A
P_A = casos_A / total_casos

# Probabilidad de B
P_B = casos_B / total_casos

# Probabilidad de A ∪ B (Unión de A y B)
P_A_union_B = (P_A + P_B - casos_A_interseccion_B / total_casos)

# Probabilidad de A^c (complemento de A)
P_A_complemento = 1 - P_A

# Mostrar resultados
print(f"P(A) = {P_A} = {casos_A}/{total_casos}")
print(f"P(B) = {P_B} = {casos_B}/{total_casos}")
print(f"P(A ∪ B) = {P_A_union_B} = 19/36")
print(f"P(A^c) = {P_A_complemento} = {11}/{12}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular la probabilidad de que n personas tengan cumpleaños diferentes
def probabilidad_diferentes(n):
    probabilidad = 1.0
    for k in range(n):
        probabilidad *= (365 - k) / 365
    return probabilidad

# Valores de n de 1 a 80
n_values = np.arange(1, 81)

# Calcular la probabilidad para cada valor de n
probabilidades = [probabilidad_diferentes(n) for n in n_values]

# Graficar la probabilidad
plt.plot(n_values, probabilidades, label="Probabilidad de cumpleaños diferentes")
plt.xlabel("Número de personas (n)")
plt.ylabel("Probabilidad")
plt.title("Probabilidad de que n personas tengan fechas de cumpleaños diferentes")
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.show()
```

```python
import numpy as np

# Función para calcular el número total de resultados posibles (36)
def total_resultados():
    return 36  # 6 caras en el primer dado * 6 caras en el segundo dado

# Evento A: La suma de los dos dados es 8
def evento_A():
    # Resultados favorables: (2,6), (3,5), (4,4), (5,3), (6,2)
    return [(2, 6), (3, 5), (4, 4), (5, 3), (6, 2)]

# Evento B: El segundo dado es impar
def evento_B():
    # El segundo dado puede ser 1, 3, o 5
    # Para cada valor del primer dado (1 a 6), hay 3 posibles resultados en los que el segundo dado es impar
    resultados = []
    for i in range(1, 7):  # Primer dado de 1 a 6
        for j in [1, 3, 5]:  # Segundo dado impar
            resultados.append((i, j))
    return resultados

# Calcular la probabilidad de un evento dado sus resultados favorables
def calcular_probabilidad(evento, total_resultados):
    return len(evento) / total_resultados

# Evento A ∩ B: La suma es 8 y el segundo dado es impar
def evento_A_interseccion_B():
    # De los resultados de la suma 8, tomamos los que tienen el segundo dado impar: (3,5), (5,3)
    return [(3, 5), (5, 3)]

# Cálculo de las probabilidades
total = total_resultados()

# Probabilidad de A: La suma es 8
A = evento_A()
P_A = calcular_probabilidad(A, total)

# Probabilidad de B: El segundo dado es impar
B = evento_B()
P_B = calcular_probabilidad(B, total)

# Probabilidad de A ∩ B: La suma es 8 y el segundo dado es impar
A_interseccion_B = evento_A_interseccion_B()
P_A_interseccion_B = calcular_probabilidad(A_interseccion_B, total)

# Producto de las probabilidades P(A) * P(B)
P_A_por_P_B = P_A * P_B

# Imprimir los resultados
print(f"Probabilidad de que la suma sea 8 (P(A)): {P_A:.4f}")
print(f"Probabilidad de que el segundo dado sea impar (P(B)): {P_B:.4f}")
print(f"Probabilidad de que la suma sea 8 y el segundo dado sea impar (P(A ∩ B)): {P_A_interseccion_B:.4f}")
print(f"Producto de las probabilidades P(A) * P(B): {P_A_por_P_B:.4f}")

# Conclusión
if np.isclose(P_A_interseccion_B, P_A_por_P_B):
    print("Los eventos A y B son independientes.")
else:
    print("Los eventos A y B no son independientes.")
```

```python
import numpy as np

# Número de experimentos (lanzamientos de 3 dados)
N = 10**5

# Función para simular el lanzamiento de 3 dados
def lanzar_tres_dados():
    return np.random.randint(1, 7, size=3)  # Genera tres números aleatorios entre 1 y 6

# Contador para los eventos favorables (exactamente un par)
contar_un_par = 0

# Realizar N experimentos
for _ in range(N):
    dados = lanzar_tres_dados()
    # Verificar si hay exactamente un par
    if len(set(dados)) == 2:  # Si hay exactamente 2 valores únicos, es un par
        contar_un_par += 1

# Calcular la probabilidad empírica
probabilidad_empirica = contar_un_par / N
print(f"Probabilidad empírica de obtener exactamente un par: {probabilidad_empirica:.4f}")
```

```python
import numpy as np

# Número de experimentos (lanzamientos de 5 dados)
N = 10**5

# Función para simular el lanzamiento de 5 dados
def lanzar_cinco_dados():
    return np.random.randint(1, 7, size=5)  # Genera 5 números aleatorios entre 1 y 6

# Contadores para los eventos
contar_un_par = 0
contar_dos_pares = 0
contar_cuatro_iguales = 0

# Realizar N experimentos
for _ in range(N):
    dados = lanzar_cinco_dados()
    # Contar las ocurrencias de cada número en los 5 dados
    cuenta = np.bincount(dados - 1)  # Restamos 1 para que 1 se corresponda con el índice 0

    # Verificar el evento de 1 par
    if np.count_nonzero(cuenta == 2) == 1 and np.count_nonzero(cuenta == 1) == 3:
        contar_un_par += 1

    # Verificar el evento de 2 pares distintos
    if np.count_nonzero(cuenta == 2) == 2 and np.count_nonzero(cuenta == 1) == 1:
        contar_dos_pares += 1

    # Verificar el evento de 4 iguales
    if np.count_nonzero(cuenta == 4) == 1 and np.count_nonzero(cuenta == 1) == 1:
        contar_cuatro_iguales += 1

# Calcular las probabilidades empíricas
P_A_empirica = contar_un_par / N
P_B_empirica = contar_dos_pares / N
P_C_empirica = contar_cuatro_iguales / N

# Mostrar los resultados
print(f"Probabilidad empírica de obtener 1 par: {P_A_empirica:.4f}")
print(f"Probabilidad empírica de obtener 2 pares distintos: {P_B_empirica:.4f}")
print(f"Probabilidad empírica de obtener 4 iguales: {P_C_empirica:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de transición
P = np.array([
    [0.30, 0.25, 0.15, 0.10],
    [0.30, 0.25, 0.55, 0.20],
    [0.30, 0.50, 0.10, 0.30],
    [0.10, 0.00, 0.20, 0.40]
])

# Evolución para n pasos
n_steps = 10
initial_probabilities = np.array([1, 0, 0, 0])  # Estado inicial
evolution = [initial_probabilities @ np.linalg.matrix_power(P, n) for n in range(n_steps + 1)]
evolution = np.array(evolution)

# Crear la gráfica
states = ["Estado 1", "Estado 2", "Estado 3", "Estado 4"]
plt.figure(figsize=(10, 6))
for i, state in enumerate(states):
    plt.plot(range(n_steps + 1), evolution[:, i], marker='o', label=state)

# Configuración del gráfico
plt.title("Evolución de las Probabilidades de Transición por Estado", fontsize=14)
plt.xlabel("Pasos (n)", fontsize=12)
plt.ylabel("Probabilidad", fontsize=12)
plt.ylim(0, 1)
plt.xticks(range(n_steps + 1), fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend(title="Estados", fontsize=10, loc='upper right')

# Mostrar gráfico
plt.tight_layout()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

# Definición de parámetros del modelo
# Matriz de transición (T)
T = np.array([[0.8, 0.2],
              [0.2, 0.8]])

# Matriz de emisión (E)
E = np.array([[0.5, 0.9],  # C: Cara
              [0.5, 0.1]])  # S: Sello

# Distribución a priori (π)
pi = np.array([0.2, 0.8])  # [P(J), P(B)]

# Secuencia observada (ΩO): S = 1, C = 0
observations = np.array([1, 0, 0, 0, 1, 0, 1, 0])

# Algoritmo hacia adelante (Forward Algorithm)
def forward(T, E, pi, observations):
    n_states = T.shape[0]
    n_observations = len(observations)

    # Inicializar matriz alpha
    alpha = np.zeros((n_states, n_observations))

    # Paso de inicialización
    alpha[:, 0] = pi * E[observations[0], :]

    # Paso de inducción
    for t in range(1, n_observations):
        for j in range(n_states):
            alpha[j, t] = E[observations[t], j] * np.sum(alpha[:, t-1] * T[:, j])

    # Probabilidad total de la secuencia observada
    P_O = np.sum(alpha[:, -1])

    return alpha, P_O

# Numeral a) Probabilidad total con distribución inicial π = [0.2, 0.8]
alpha, P_O = forward(T, E, pi, observations)
print(f"Probabilidad total de la secuencia observada con π = [0.2, 0.8]: {P_O}")

# Numeral b) Secuencia oculta más probable (Viterbi Algorithm)
def viterbi(T, E, pi, observations):
    n_states = T.shape[0]
    n_observations = len(observations)

    # Matrices de Viterbi
    delta = np.zeros((n_states, n_observations))  # Máxima probabilidad
    psi = np.zeros((n_states, n_observations), dtype=int)  # Punteros para backtracking

    # Inicialización
    delta[:, 0] = pi * E[observations[0], :]

    # Recurrencia
    for t in range(1, n_observations):
        for j in range(n_states):
            probs = delta[:, t-1] * T[:, j]
            delta[j, t] = np.max(probs) * E[observations[t], j]
            psi[j, t] = np.argmax(probs)

    # Terminación y backtracking
    states = np.zeros(n_observations, dtype=int)
    states[-1] = np.argmax(delta[:, -1])
    for t in range(n_observations-2, -1, -1):
        states[t] = psi[states[t+1], t+1]

    P_max = np.max(delta[:, -1])
    return states, P_max

hidden_states, P_max = viterbi(T, E, pi, observations)
print(f"Secuencia oculta más probable: {hidden_states}")
print(f"Probabilidad de la secuencia más probable: {P_max}")

# Numeral c) Probabilidades de los estados observables
observable_probs = np.sum(alpha, axis=0)  # Suma de probabilidades por estado observable
sum_observable_probs = np.sum(observable_probs)
print(f"Suma de probabilidades de los estados observables: {sum_observable_probs}")

# Numeral d) Verificación de ∑Po = 1 (Gráfica)

# Algoritmo hacia adelante (Forward Algorithm)
alpha, P_O = forward(T, E, pi, observations)




# Numeral e) Probabilidades de todas las secuencias ocultas (Figura 6.9)
def sequence_probability(hidden_seq, observations, T, E, pi):
    prob = pi[hidden_seq[0]] * E[observations[0], hidden_seq[0]]  # Inicialización
    for t in range(1, len(hidden_seq)):
        prob *= T[hidden_seq[t-1], hidden_seq[t]] * E[observations[t], hidden_seq[t]]  # Recurrencia
    return prob

# Enumerar todas las posibles secuencias ocultas (256 combinaciones para 8 lanzamientos)
hidden_states_all = list(product([0, 1], repeat=len(observations)))

# Calcular las probabilidades para todas las secuencias ocultas
sequence_probs = [sequence_probability(seq, observations, T, E, pi) for seq in hidden_states_all]

# Graficar las probabilidades de las secuencias ocultas
plt.figure(figsize=(12, 6))
plt.bar(range(len(sequence_probs)), sequence_probs, color='skyblue', edgecolor='black')
plt.title('Probabilidades de Cada Secuencia Oculta para el Estado Observado (Figura 6.9)')
plt.xlabel('Índice de Secuencia Oculta')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Verificar suma de las probabilidades
total_sequence_prob = sum(sequence_probs)
print(f"Suma total de las probabilidades de todas las secuencias ocultas: {total_sequence_prob}")

```

```python
#Punto 10
def velocidad_media(d1, d2, d3, v1, v2, v3):
    # Fórmula de la velocidad media en un viaje con diferentes distancias y velocidades
    tiempo_total = (d1 / v1) + (d2 / v2) + (d3 / v3)
    distancia_total = d1 + d2 + d3
    velocidad_media = distancia_total / tiempo_total
    return velocidad_media

# Distancias (en millas) y velocidades (en millas por hora)
d1 = 100  # distancia del tramo 1
d2 = 150  # distancia del tramo 2
d3 = 200  # distancia del tramo 3

v1 = 200  # velocidad en el tramo 1 (millas por hora)
v2 = 150  # velocidad en el tramo 2 (millas por hora)
v3 = 100  # velocidad en el tramo 3 (millas por hora)

# Calcular la velocidad media
v_media = velocidad_media(d1, d2, d3, v1, v2, v3)

# Mostrar el resultado
print(f"La velocidad media del avión en el viaje completo es: {v_media:.2f} millas por hora.")
```

```python
#Punto 9
import numpy as np
import matplotlib.pyplot as plt

# Definir la función Beta f(x; 2, 4)
def f(x, alpha=2, beta=4):
    # Gamma(n) se aproxima por factorial en este caso
    return (np.math.gamma(alpha + beta) / (np.math.gamma(alpha) * np.math.gamma(beta))) * x**(alpha - 1) * (1 - x)**(beta - 1)

# Función para el método de aceptación y rechazo
def metodo_aceptacion_rechazo(n_muestras, alpha=2, beta=4):
    # Definir la constante M
    M = f(0.5, alpha, beta)  # Aproximación de M, ya que la Beta tiene un máximo en x=0.5 para alpha=2 y beta=4

    # Muestras aceptadas
    aceptadas = 0

    # Generar n_muestras puntos
    for _ in range(n_muestras):
        # Generamos un punto de la distribución uniforme en [0, 1]
        x = np.random.uniform(0, 1)
        # Generamos un número aleatorio u en [0, 1]
        u = np.random.uniform(0, 1)

        # Verificamos si aceptamos o rechazamos la muestra
        if u <= f(x, alpha, beta) / M:
            aceptadas += 1

    # Estimación del área (probabilidad) de aceptación
    area_estimada = aceptadas / n_muestras
    return area_estimada

# Número de muestras
n_muestras = 100000
area_estimacion = metodo_aceptacion_rechazo(n_muestras)

# Mostrar el resultado
print(f"Área estimada bajo la curva: {area_estimacion:.4f}")

# Estimación de la incertidumbre
incertidumbre = np.sqrt(area_estimacion * (1 - area_estimacion) / n_muestras)
print(f"Incertidumbre estimada: {incertidumbre:.4f}")

# Mostrar la curva Beta para visualización
x = np.linspace(0, 1, 1000)
y = [f(i) for i in x]
plt.plot(x, y, label="f(x; 2, 4)", color="blue")
plt.fill(x, y, color="blue", alpha=0.3)

plt.title("Distribución Beta f(x; 2, 4) con Método de Aceptación y Rechazo")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
