```python
import numpy as np

# Función de likelihood (definir como función separada)
def likelihood(mu, epsilon, data, background, signal):
    # Implementar la función de likelihood como discutido antes
    # Devolver el valor de likelihood para un (mu, epsilon)
    # Example likelihood function (replace with your actual function)
    likelihood_value = 1.0  # Initialize likelihood
    for i in range(len(data)):
        expected_counts = background[i] + mu * epsilon * signal[i]
        likelihood_value *= np.exp(-expected_counts) * (expected_counts ** data[i]) / np.math.factorial(data[i])
    return likelihood_value

# Inicialización
mu = 1.5
epsilon = 1.0
n_iterations = 10000
sigma = 0.1
samples = []

data = [10, 20, 5]
background = [6, 15, 10]
signal = [2, 10, 3]

for t in range(n_iterations):
    # Propuesta
    mu_prime = np.random.normal(mu, sigma)
    epsilon_prime = np.random.normal(epsilon, sigma)

    # Asegurar que los nuevos valores están en los límites
    if mu_prime < 0:
        mu_prime = mu  # Rechazar
    if epsilon_prime < 0.5 or epsilon_prime > 1.5:
        epsilon_prime = epsilon  # Rechazar

    # Calcular likelihoods
    L_current = likelihood(mu, epsilon, data, background, signal)
    L_proposed = likelihood(mu_prime, epsilon_prime, data, background, signal)

    # Criterio de aceptación
    r = L_proposed / L_current

    if r >= 1 or np.random.uniform(0, 1) < r:
        mu, epsilon = mu_prime, epsilon_prime  # Aceptar
    samples.append((mu, epsilon))

# Convertir muestras a un array para análisis
samples = np.array(samples)

# Estimar parámetros
mu_estimate = np.mean(samples[:, 0])
epsilon_estimate = np.mean(samples[:, 1])

# Calcular errores estándar
mu_error = np.std(samples[:, 0])
epsilon_error = np.std(samples[:, 1])

print(f"Estimación de μ: {mu_estimate} ± {mu_error}")
print(f"Estimación de ε: {epsilon_estimate} ± {epsilon_error}")
```

A partir de la estimacio relizada se puede decir que no se puede descartar la pocibilidad de vida extraterrestre

```python
import numpy as np
import random

def LennardJones(r, epsilon=1, sigma=1):
    """
    Potencial de Lennard-Jones para dos partículas a una distancia r.
    """
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

def TotalEnergy(positions, epsilon=1, sigma=1):
    """
    Calcula la energía total de un sistema de partículas en posiciones dadas.
    """
    N = len(positions)
    energy = 0
    for i in range(N):
        for j in range(i + 1, N):
            r = np.linalg.norm(positions[i] - positions[j])  # Distancia entre partículas
            if r != 0:
                energy += LennardJones(r, epsilon, sigma)
    return energy

def generate_population(pop_size, N):
    """
    Genera una población inicial de posiciones aleatorias para N partículas.
    """
    return [np.random.rand(N, 3) for _ in range(pop_size)]

def select_parents(population, energies):
    """
    Selecciona dos padres basándose en su energía (menor es mejor).
    """
    indices = np.argsort(energies)[:2]  # Selecciona los dos de menor energía
    return population[indices[0]], population[indices[1]]

def crossover(parent1, parent2):
    """
    Realiza el cruce entre dos padres para generar un nuevo individuo.
    """
    child = np.copy(parent1)
    for i in range(len(child)):
        if random.random() < 0.5:
            child[i] = parent2[i]
    return child

def mutate(individual, mutation_rate=0.01):
    """
    Aplica mutaciones aleatorias a un individuo.
    """
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            individual[i] += np.random.normal(0, 0.1, size=3)  # Mutación gaussiana
    return individual

def genetic_algorithm(pop_size=100, generations=250, N=50):
    # Inicializar la población
    population = generate_population(pop_size, N)

    for generation in range(generations):
        energies = [TotalEnergy(individual) for individual in population]
        new_population = []

        for _ in range(pop_size // 2):
            parent1, parent2 = select_parents(population, energies)
            child1 = crossover(parent1, parent2)
            child2 = crossover(parent2, parent1)
            new_population.append(mutate(child1))
            new_population.append(mutate(child2))

        population = new_population

        # Mostrar el mejor individuo de cada generación
        best_energy = min(energies)
        print(f"Generación {generation + 1}: Energía mínima = {best_energy:.4f}")

    # Encontrar el mejor individuo final
    final_energies = [TotalEnergy(individual) for individual in population]
    best_index = np.argmin(final_energies)
    best_positions = population[best_index]
    best_energy = final_energies[best_index]

    return best_positions, best_energy

# Ejecución del algoritmo genético
best_positions, best_energy = genetic_algorithm()
print(f"\nMejores posiciones:\n{best_positions}")
print(f"Energia total mínima: {best_energy:.4f}")
```
