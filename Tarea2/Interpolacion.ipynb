```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
x = sym.Symbol('x')
```

```python
# prompt: Genera un par X, Y con 5 puntos y que sea un array

X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 1, 3, 5])
```

```python
def Lagrange(x, X, i):
  L=1
  for j in range(len(X)):
    if j!=i:
      L*= (x-X[j])/(X[i]-X[j])
  return L
```

```python
Lagrange(x, X, 0)
```

```python
def NewtonRapson(x,X,Y):
  L = 0
  for i in range(len(X)):
    L += Y[i]*Lagrange(x,X,i)
  return L
```

```python
plt.plot(X,Y, 'o')
```

```python
x1 = np.linspace(X[0], X[-1], 50)
plt.plot(X, Y, "o")
plt.plot(x1, NewtonRapson(x1, X, Y))
```
