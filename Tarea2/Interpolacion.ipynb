```python
# prompt: Genera un par X, Y con 5 puntos y que sea un array

X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 1, 3, 5])
```

L de Lagrange

```python
def Lagrange(x, X, i):
  L=1
  for j in range(len(X)):
    if j!=i:
      L*= (x-X[j])/(X[i]-X[j])
  return L
```

```python
Lagrange(x, X, 0)
```

Interpolación de lagrange

```python
def Interpolate(x,X,Y):
  L = 0
  for i in range(len(X)):
    L += Y[i]*Lagrange(x,X,i)
  return L
```

```python
plt.plot(X,Y, 'o')
```

```python
x1 = np.linspace(X[0], X[-1], 50)
plt.plot(X, Y, "o")
plt.plot(x1, Interpolate(x1, X, Y))
```

```python
def f(x):
    return np.sqrt(np.tan(x))
```

```python
def f_prima(x):
    exacta = (1)/(2*((np.sqrt(np.tan(x)))*(np.cos(x)**2)))
    return exacta
```

```python
def ProgressiveDerivate(f, x, h=0.01):
    return (f(x+h)-f(x))/h
def CentralDerivate(f, x, h=0.01):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
X = np.linspace(1.0, 1.1)
Y = f(X)
prima = f_prima(X)
Y_central = CentralDerivate(f, X)
Y_progressive = ProgressiveDerivate(f, X)
error_central = np.abs(prima-Y_central)
error_progressive = np.abs(prima-Y_progressive)
```

```python
plt.scatter(X, Y, color="b")
plt.scatter(X, prima, color="k")
plt.scatter(X, Y_central, color="r")
plt.scatter(X, Y_progressive, color="g")
```

```python
plt.plot(X, error_central, color="r")
plt.plot(X, error_progressive, color="g")
```
